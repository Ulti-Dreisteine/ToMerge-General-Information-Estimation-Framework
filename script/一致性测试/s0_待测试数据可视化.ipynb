```python
import numpy as np
import sys
import os

BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), "../" * 2))
sys.path.insert(0, BASE_DIR)

from setting import plt
from dataset.bivariate.data_generator import DataGenerator
```

```python
funcs = [
    # "circular",
    "sin_high_freq", 
    "cos_high_freq", 
    "vary_freq_sin", 
    "vary_freq_cos", 
    "non_fourier_freq_sin",
    "non_fourier_freq_cos", 
    "sin_low_freq", 
    "linear_periodic_med_freq",
    "cubic", 
    "cubic_y_stretched", 
    "parabola", 
    "spike",
    "lopsided_l_shaped", 
    "l_shaped", 
    "sigmoid",
    "exp_base_10",
    "exp_base_2",
    "linear_periodic_high_freq_2", 
    "linear_periodic_high_freq",
    "linear_periodic_low_freq",
    "line",
]

funcs_alias = [
    # "circular",
    "sin high freq", 
    "cos high freq", 
    "vary freq sin", 
    "vary freq cos", 
    "non fourier freq sin",
    "non fourier freq cos", 
    "sin low freq", 
    "linear periodic med freq",
    "cubic", 
    "cubic y stretched", 
    "parabola", 
    "spike",
    "lopsided l shaped", 
    "l shaped", 
    "sigmoid",
    "exp base 10",
    "exp base 2",
    # "linear periodic high freq 2", 
    "linear periodic high freq",
    "linear periodic low freq",
    "line",
]
```

```python
N = 1000
data_gener = DataGenerator()

fig, axs = plt.subplots(len(funcs) // 7, 7, figsize=[13, 5])
for i, func in enumerate(funcs):
    ax = axs[i // 7, i % 7]
    x, y, _, _ = data_gener.gen_data(N, func, normalize=False)
    if func != "circular":
        sort_idxs = np.argsort(x)
        x, y = x[sort_idxs], y[sort_idxs]
    ax.plot(x, y, linewidth=2)
    # ax.set_title(funcs_alias[i], fontsize = 12)
    ax.set_title(funcs[i], fontsize = 12)
    ax.get_yaxis().get_major_formatter().set_scientific(True)

    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")

fig.tight_layout()
```
